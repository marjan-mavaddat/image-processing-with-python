```python
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
img = io.imread('/content/drive/MyDrive/imageprocessing/lena.JPG', as_gray=True) * 255
power = img.copy()
for x in range(0, power.shape[0]):
  for y in range(0, power.shape[1]):
    power[x, y] = power[x, y] ** 2
    if power[x, y] >= 255:
      power[x, y] = 255
power = np.array(power, dtype=np.uint8)
f, axarr = plt.subplots(1, 2)
axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(power, cmap='gray')
axarr[0].title.set_text("Original")
axarr[1].title.set_text("Nth Power")
plt.show()
```

**نمایش هیستوگرام تصویر رنگی**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
  
img = cv2.imread('/content/drive/MyDrive/imageprocessing/lena.JPG')
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Calculate histogram without mask
hist1 = cv2.calcHist([image],[0],None,[256],[0,256])
hist2 = cv2.calcHist([image],[1],None,[256],[0,256])
hist3 = cv2.calcHist([image],[2],None,[256],[0,256])

plt.subplot(221), plt.imshow(image)
plt.subplot(222), plt.plot(hist1, color = "blue"), plt.plot(hist2, color = "green"),plt.plot(hist3, color = "red"),plt.grid()

plt.title("histogram image color")
plt.xlabel("Intensity Value")
plt.ylabel("Count")
plt.legend(['Blue', 'Green', 'Red'])
plt.xlim([0,256])
```

**اعمال فیلتر برتصاویر رنگی**

**1. Sharpening an Image**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Loading source image
src_image = cv2.imread('/content/drive/MyDrive/imageprocessing/lena.JPG')
image = cv2.cvtColor(src_image, cv2.COLOR_BGR2RGB)
# Defining the kernel of size 3x3
kernel = np.array([
  [0, -1, 0],
  [-1, 5, -1],
  [0, -1, 0]
])
 
resulting_image = cv2.filter2D(image, -1, kernel)

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("original image")

plt.subplot(1,2,2)
plt.imshow(resulting_image)
plt.title("Sharpened Image")
```

**2. Bluring an image**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Loading source image
src_image = cv2.imread('/content/drive/MyDrive/imageprocessing/lena.JPG')
image = cv2.cvtColor(src_image, cv2.COLOR_BGR2RGB)
# Defining the kernel of size 3x3
kernel = np.array([
  [1, 1, 1],
  [1, 1, 1],
  [1, 1, 1]
]) / 9
 
resulting_image = cv2.filter2D(image, -1, kernel)

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("original image")

plt.subplot(1,2,2)
plt.imshow(resulting_image)
plt.title("Bluring Image")
```

** فیلتر بدون استفاده از تابع  **

```python
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
img = io.imread('/content/drive/MyDrive/imageprocessing/lena.JPG', as_gray=True)
mask = np.ones((3, 3), dtype=int)
mask = mask / 9
# Convolve the 3X3 mask over the image
average = np.zeros([img.shape[0], img.shape[1]])
for i in range(1, img.shape[0] - 1):
  for j in range(1, img.shape[1] - 1):
    temp = img[i - 1, j - 1] * mask[0, 0] + img[i - 1, j] * mask[0, 1] + img[i- 1, j + 1] * mask[0, 2] 
    + img[i, j - 1] * mask[1, 0] + img[i, j] * mask[1, 1] + img[i, j + 1] * mask[1, 2] 
    + img[i + 1, j - 1] * mask[2, 0] + img[i + 1, j] * mask[2, 1] + img[i + 1, j + 1] * mask[2, 2]
    average[i, j] = temp
f, axarr = plt.subplots(1, 2)
axarr[0].imshow(img, cmap='gray')
axarr[1].imshow(average, cmap='gray')
axarr[0].title.set_text("Original")
axarr[1].title.set_text("Average 3x3")
plt.show()
```

**افزودن روشنایی تصویر**

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread('/content/drive/MyDrive/imageprocessing/lena.JPG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
r = img.copy()

for i in range(r.shape[0]):
  for j in range(r.shape[1]):
    for k in range(r.shape[2]):
      r[i, j, k] = np.clip(r[i, j, k] + 128 , 0, 255 )

f, axarr = plt.subplots(1, 2)
axarr[0].imshow(img)
axarr[1].imshow(r)
axarr[0].title.set_text("Original")
axarr[1].title.set_text("RGB+128")
plt.show()
```

**افزودن روشنایی تصویر به طریقی دیگر**

```python
import pygame
import cv2
import matplotlib.pyplot as plt
img = pygame.image.load('/content/drive/MyDrive/imageprocessing/lena.JPG')
W, H = img.get_width(), img.get_height()
f = 128

def fix_color(color):
    if color > 255:
        return 255
    elif color < 0:
        return 0
    else:
        return int(color)

for x in range(W):
    for y in range(H):
        color = pygame.Surface.get_at(img, (x,y))
        R , G , B = (color.r +f) , (color.g +f) , (color.b +f)
        R , G , B = fix_color(R), fix_color(G), fix_color(B)
        color.update(R,G,B)
        pygame.Surface.set_at(img, (x,y), color)

pygame.image.save(img, '/content/drive/MyDrive/imageprocessing/lena2.JPG')

path='/content/drive/MyDrive/imageprocessing/lena2.JPG'
img2 = cv2.imread(path)
image2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

path2='/content/drive/MyDrive/imageprocessing/lena.JPG'
img3 = cv2.imread(path2)
image3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

plt.imshow(image3)
plt.show()
plt.imshow(image2)
plt.show()
```
